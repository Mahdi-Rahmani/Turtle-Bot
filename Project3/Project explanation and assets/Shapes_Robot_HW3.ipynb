```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Rectangle

X1 = np.linspace(-3, 3 , 100)
Y1 = np.array([2]*100)


Y2 = np.linspace(2, -2 , 100)
X2 = np.array([3]*100)

X3 = np.linspace(3, -3 , 100)
Y3 = np.array([-2]*100)

Y4 = np.linspace(-2, 2 , 100)
X4 = np.array([-3]*100)


plt.plot(np.concatenate([X1,X2, X3 , X4]), np.concatenate([Y1,Y2,Y3,Y4]))
plt.show()
```

```python
# Octagon

X1 = np.linspace(-1, 1 , 100)
Y1 = np.array([3]*100)

X2 = np.linspace(1, 1 + 2**(1/2) , 100)
Y2 = - (2**(1/2)) * (X2 - 1) + 3

Y3 = np.linspace(1, -1 , 100)
X3 = np.array([1 + 2**(1/2)]*100)

X4 = np.linspace(1 + 2**(1/2), 1, 100)
Y4 = (2**(1/2)) * (X4 - 1 - 2**(1/2)) -1 

X5 = np.linspace(1, -1 , 100)
Y5 = np.array([-3]*100)

X6 = np.linspace(-1, -1 - 2**(1/2) , 100)
Y6 = - (2**(1/2)) * (X6 + 1) - 3 


Y7 = np.linspace(-1, 1 , 100)
X7 = np.array([- 1 - 2**(1/2)]*100)


X8 = np.linspace(-1 - 2**(1/2), -1, 100)
Y8 = (2**(1/2)) * (X8 + 1 + 2**(1/2)) + 1


plt.plot(np.concatenate([X1,X2,X3,X4,X5,X6,X7,X8]), np.concatenate([Y1,Y2,Y3,Y4,Y5,Y6,Y7,Y8]))
plt.show()
```

```python
# Star

X1 = np.linspace(0, 3 , 100)
Y1 = - (7/3) * X1  + 12

X2 = np.linspace(3, 10 , 100)
Y2 = np.array([5]*100)

X3 = np.linspace(10, 4 , 100)
Y3 = (5/6) * X3  - (10/3)

X4 = np.linspace(4, 7 , 100)
Y4 = -(3) * X4  + 12

X5 = np.linspace(7, 0 , 100)
Y5 = -(4/7) * X5  - 5

X6 = np.linspace(0, -7 , 100)
Y6 = (4/7) * X6  - 5

X7 = np.linspace(-7, -4 , 100)
Y7 = 3 * X7  + 12

X8 = np.linspace(-4, -10 , 100)
Y8 = -(5/6) * X8  - (10/3)

X9 = np.linspace(-10, -3 , 100)
Y9 = np.array([5]*100)

X10 = np.linspace(-3, 0 , 100)
Y10 = (7/3) * X10  + 12


plt.plot(np.concatenate([X1, X2, X3, X4, X5, X6, X7, X8, X9, X10]), np.concatenate([Y1, Y2, Y3, Y4, Y5, Y6, Y7, Y8, Y9, Y10]))
plt.show()
```

```python
# Heart

x_dim, y_dim = 4,4

t = np.linspace(0, np.pi*(2/3), 100)
X1 = x_dim * np.cos(t) + 2
Y1 = y_dim * np.sin(t)

t = np.linspace(np.pi*(-1), np.pi*(-5/3), 100)
X2 = x_dim * np.cos(t) -2
X2 = np.flip(X2)
Y2 = y_dim * np.sin(t)
Y2 = np.flip(Y2)

X3 = np.linspace(-6, 0 , 100)
Y3 = (-4/3) * X3  - 8

X4 = np.linspace(0, 6 , 100)
Y4 = (4/3) * X4  - 8


plt.plot(np.concatenate([X1, X2, X3, X4]), np.concatenate([Y1, Y2, Y3, Y4]))
plt.show()
```

```python
# Logarithmic spiral 

a = 0.17
k = math.tan(a)
X , Y = [] , []

for i in range(150):
    t = i / 20 * math.pi
    dx = a * math.exp(k * t) * math.cos(t)
    dy = a * math.exp(k * t) * math.sin(t)
    X.append(dx)
    Y.append(dy) 
plt.plot(X,Y)
plt.show()
```
